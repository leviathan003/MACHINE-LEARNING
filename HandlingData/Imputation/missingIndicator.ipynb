Using a seperate col and using values of TRUE where the corresponding data is missing and FALSE where corresponding data is present

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import MissingIndicator,SimpleImputer
```

```python
df = pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])
df.head()
```

```python
X = df.drop(columns=['Survived'])
Y = df['Survived']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

Without Missing Indicator

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

clf = LogisticRegression()
clf.fit(X_train_trf,Y_train)

y_pred = clf.predict(X_test_trf)

accuracy_score(Y_test,y_pred)
```

With Missing Indicator

```python
mi = MissingIndicator()
mi.fit(X_train)
X_train_missing = mi.transform(X_train)
X_test_missing = mi.transform(X_test)
X_train['Age_NA'] = X_train_missing
X_test['Age_NA'] = X_test_missing
```

```python
si2 = SimpleImputer()
X_train_trf2 = si2.fit_transform(X_train)
X_test_trf2 = si2.fit_transform(X_test)
```

```python
clf.fit(X_train_trf2,Y_train)

y_pred = clf.predict(X_test_trf2)

accuracy_score(Y_test,y_pred)
```

```python
# directly use the feature in Simple imputer as follows
si = SimpleImputer(add_indicator=True)

X_train = si.fit_transform(X_train)
X_test = si.fit_transform(X_test)

clf.fit(X_train,Y_train)
y_pred = clf.predict(X_test)
accuracy_score(Y_test,y_pred)
```

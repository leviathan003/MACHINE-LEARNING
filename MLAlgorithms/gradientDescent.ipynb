```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,Y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=100,random_state=13)
plt.scatter(X,Y)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,Y)
```

```python
reg.coef_,reg.intercept_
```

```python
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red')
```

lets apply gradient descent assuming m(slope) is constant at 82.71 and b(intercept) = 0

```python
y_pred = ((82.71*X)+0).reshape(4)
```

```python
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

1st Iteration

```python
m=82.71
b=0

loss_slope = -2*np.sum(Y-m*X.ravel()-b)
loss_slope
```

```python
learning_rate = 0.1
step_size = loss_slope*learning_rate
step_size
```

new intercept

```python
b = b-step_size
b1=b
b
```

```python
y_pred1 = ((82.71*X)+b).reshape(4)
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0',alpha=0.15)
plt.plot(X,y_pred1,color='blue',label='b={}'.format(b1))
plt.legend()
plt.show()
```

2nd Iteration

```python
loss_slope = -2*np.sum(Y-m*X.ravel() -b)
step_size = loss_slope*learning_rate
b = b-step_size
b2=b
loss_slope,step_size,b
```

```python
y_pred2 = ((82.71*X)+b).reshape(4)
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0',alpha=0.13)
plt.plot(X,y_pred1,color='blue',label='b={}'.format(b1),alpha=0.15)
plt.plot(X,y_pred2,color='orange',label='b={}'.format(b2))
plt.legend()
plt.show()
```

3rd Iteration

```python
loss_slope = -2*np.sum(Y-m**X.ravel()-b)
step_size = loss_slope*learning_rate
b=b-step_size
b3=b
loss_slope,step_size,b
```

```python
y_pred2 = ((82.71*X)+b).reshape(4)
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0',alpha=0.13)
plt.plot(X,y_pred1,color='blue',label='b={}'.format(b1),alpha=0.15)
plt.plot(X,y_pred2,color='orange',label='b={}'.format(b2),alpha=0.18)
plt.plot(X,y_pred2,color='orange',label='b={}'.format(b3))
plt.legend()
plt.show()
```

The line at 3rd iteration matches that of the actual line so we achieved the est fit line via Gradient Descent

Summarizing using a loop

```python
b = -100
m = 82.71
learning_rate=0.1
epochs=10

for i in range(epochs):
    loss_slope = -2*np.sum(Y - m*X.ravel() -b)
    step_size = loss_slope*learning_rate
    b = b-step_size

    y_pred = m*X + b

    plt.plot(X,y_pred)

plt.scatter(X,Y)
```
